```python
%matplotlib inline
from pyVPRM.lib.fancy_plot import *
import xarray as xr
```

```python
idata = xr.open_dataarray('./out/nee_h18v04_2022_160.h5')
```

# NEE Plots

```python
for i in idata.coords['time']:
    fig, ax = newfig(0.9)
    idata.sel({'time': i}).plot.imshow(ax=ax, cmap='PRGn_r',
                                       vmin=-30, vmax=30, 
                                       cbar_kwargs={'label': r'NEE [$\mu\,$mol$\,$CO$_{2}\,$m$^{-2}\,$s$^{-1}$]'})
    title = str(i.values).split('T')[0] + ' ' + str(i.values).split('T')[1].split('.')[0] + ' UTC ' 
    ax.set_title(title)
    fig.show()
```
